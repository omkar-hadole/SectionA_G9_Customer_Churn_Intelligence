This notebook performs statistical analysis on the cleaned dataset to validate the patterns identified during Exploratory Data Analysis (EDA). It includes descriptive statistics, correlation analysis, and hypothesis testing to understand the factors influencing customer churn.

```python
import pandas as pd
import numpy as np

# Load dataset (change file name if needed)
df = pd.read_csv("/Users/aaryasrivastava/Desktop/SectionA_G9_Customer_Churn_Intelligence/data/processed/cleaned_banking_churn.csv")

df.head()
```

## Data Overview

The dataset contains customer demographic, financial, and transactional information used to analyze churn behavior.

## Descriptive Statistics

This section summarizes key numerical variables such as balance and transactions.

```python
df[['current_balance','current_month_credit','current_month_debit']].describe()
```

### Observations
- The average balance shows the overall financial status of customers.
- Transaction values indicate customer activity levels.
- Variation in values suggests differences in customer behavior.

## Correlation Analysis

This section examines the relationship between variables and churn.


```python
df.corr(numeric_only=True)['churn'].sort_values(ascending=False)
```

### Observations
- Variables with higher correlation values have stronger relationships with churn.
- Financial and activity-related features may influence customer churn.
