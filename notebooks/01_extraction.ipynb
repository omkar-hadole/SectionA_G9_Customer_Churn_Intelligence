# Notebook 01: Data Extraction

In this notebook, we load the raw data and prove that it passes the **Capstone 2 Minimum Standards** (At least 5,000 rows and 8 meaningful analytical columns). We are not altering or cleaning any data here.

### Step 1: Import Libraries

```python
import pandas as pd
```

### Step 2: Load the Raw Dataset
Loading the completely untouched raw data from our `data/raw/` directory.

```python
# Load dataset
df = pd.read_csv("../data/raw/Banking_churn_prediction.csv")

# Show first 5 rows
df.head()
```

### Step 3: Check Capstone 2 Requirements
We need to prove to the evaluator that our dataset is large enough! Minimum Requirements: 5,000 rows and 8 columns.

```python
# Get number of rows and columns
rows = df.shape[0]
cols = df.shape[1]

print("Total Rows:", rows)
print("Total Columns:", cols)

# Check requirements
if rows >= 5000:
    print("Dataset meets row requirement")
else:
    print("Dataset does NOT meet row requirement")

if cols >= 8:
    print("Dataset meets column requirement")
else:
    print("Dataset does NOT meet column requirement")
```

### Step 4: Dataset Info
We investigate numeric vs categorical datatypes.

```python
# Check data types and null values
df.info()
```

### Step 5: Dataset Summary (Describe)
We look at the basic statistics (mean, min, max, etc.) for our numerical columns.

```python
# Summary of numerical columns
df.describe()
```

### Step 6: Identify Missing Values
We check exactly how many null values the ETL team needs to clean in the next step.

```python
# Count missing values in each column
df.isnull().sum()
```

### Step 7: Identify Duplicate Values
